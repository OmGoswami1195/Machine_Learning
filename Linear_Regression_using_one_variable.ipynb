# Machine Learning With Python: Linear Regression With One Variable

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head(10)
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Price in US$")
plt.scatter(df.year,df.per_capita_income_inUS,color='red',marker='+')
```

```python
df.rename(columns={'per_capita_income_inUS':'income'},inplace=True)
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df.income)
```

```python
reg.predict([[2017]])
```

```python
reg.score(df[['year']], df.income)
```

```python
plt.scatter(df.year, df.income, color='red',marker='+')
plt.xlabel('Year', fontsize=20, color='green')
plt.ylabel("Per Capita Income (Us$)", fontsize=20, color='green')
plt.plot(df.year, reg.predict(df[['year']]), color='blue')
```
